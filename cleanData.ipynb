# Question: Which phone is the most worth to buy? 
The most worth phone is defined as the cheapest phone possible which still fulfills the minimum criteria.

```python
import pandas as pd
```

```python
phoneSpecs_file = 'SpecsWebsites\phoneSpecs.csv'
phoneWeb_file = 'OfficialWebsites\OfficialWeb_Phones.csv'
```

```python
df_specs = pd.read_csv(phoneSpecs_file)
df_specs.head()
```

```python
df_web = pd.read_csv(phoneWeb_file)
df_web.head()
```

name field in both make it lower case

```python
df_specs['name'] = df_specs['name'].str.lower()
df_specs.head(2)
```

```python
df_web['name'] = df_web['name'].str.lower()
df_web.head(2)
```

name field in specs table separate from brand

```python
# get the brand to a new column
df_specs['brand'] = df_specs['name'].apply(lambda row: row.split()[0])
df_specs.head(2) # rightmost column
```

```python
# remove the brand name from the name column
try:
    df_specs['name'] = df_specs['name'].apply(lambda row: row.replace(row['brand'],'').strip()) 
except:
    pass
df_specs.head(2)
```
